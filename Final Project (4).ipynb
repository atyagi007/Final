```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
# Y = a + b*ln(X)
X = np.arange(1,50,0.5)
Y = 10 + 2*np.log(X)
Ynoise = np.random.rand(len(Y))
Y = Y +Ynoise
plt.scatter(X,Y)
c = len(X)
xbias = np.ones((c,1))
print (X.shape)
print (xbias.shape)
X = np.reshape(X,(c,1))
print (X.shape)
Xlog = np.log(X)
x_new = np.append(xbias,Xlog,axis=1)
x_new_transpose = np.transpose(x_new)
x_new_transpose_dot_x_new = x_new_transpose.dot(x_new)
temp = np.linalg.inv(x_new_transpose_dot_x_new)
temp_2 = x_new_transpose.dot(Y)
coe  = temp.dot(temp_2)
a = coe[0]
b = coe[1]
Yplot = a + b*np.log(X)
plt.scatter(X,Y)
plt.plot(X,Yplot,c="g")
Accuracy = r2_score(Y,Yplot)
print (Accuracy)
plt.title('Possible Satelite Flight Trajectories')
plt.xlabel('Horizontal distance')
plt.ylabel('height')
 
```
